```python
# 高斯法计算密度时，如果距离范围为[0,1]，阈值dc的设定对密度值的影响
import matplotlib.pyplot as plt
import numpy as np

% matplotlib inline

dc_vec = np.arange(0.1,1,0.1)

xxx = np.arange(0,1,0.01)
for dc in dc_vec:
    yyy = np.exp(-((1-xxx)/dc)**2)
    plt.plot(xxx,yyy)
```
